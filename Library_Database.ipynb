```python
class Library:
    def __init__(self, name):
        self.name = name
        self.booksList = []
        self.lended_books = {}

    def display_available_books(self):
        if len(self.booksList) == 0:
            print("No books available in the library.")
        else:
            print("Available Books:")
            for book in self.booksList:
                print(book)

    def lend_book(self, username, book):
        if book in self.booksList:
            if book not in self.lended_books.values():
                self.lended_books[username] = book
                self.booksList.remove(book)
                print(f"{book} has been lent to {username}.")
            else:
                print(f"{book} is already lended to another user.")
        else:
            print(f"Sorry, {book} is not available in the library.")

    def add_book(self, book):
        self.booksList.append(book)
        print(f"{book} has been added to the library.")

    def return_book(self, username, book):
        if username in self.lended_books and self.lended_books[username] == book:
            del self.lended_books[username]
            self.booksList.append(book)
            print(f"{book} has been returned by {username}.")
        else:
            print(f"{username} did not borrow {book} from this library or the book is not in the system.")


def main():
    my_library = Library("My Library")
    my_library.booksList = ['Lord Of The Rings', 'Rich Dad Poor Dad', 'Harry Potter', 'Percy Jackson', 'The Three Musketeers']

    print("Welcome to", my_library.name)

    while True:
        print("\nOptions:")
        print("1. Display Books")
        print("2. Lend a Book")
        print("3. Add a Book")
        print("4. Return a Book")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            my_library.display_available_books()
        elif choice == "2":
            username = input("Enter your name: ")
            book = input("Enter the name of the book you want to lend: ")
            my_library.lend_book(username, book)
        elif choice == "3":
            book = input("Enter the name of the book you want to add: ")
            my_library.add_book(book)
        elif choice == "4":
            username = input("Enter your name: ")
            book = input("Enter the name of the book you want to return: ")
            my_library.return_book(username, book)
        elif choice == "5":
            print("Thank you for using", my_library.name)
            break
        else:
            print("Invalid choice. Please enter a valid option.")


if __name__ == "__main__":
    main()
```
